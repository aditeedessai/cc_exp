a.	Python code for text handling

```python
user_input = input("Enter a string: ")

if user_input == user_input[::-1]:
    print(f"'{user_input}' is a palindrome!")
else:
    print(f"'{user_input}' is not a palindrome.")
```

b.	Python code for file handling

```python
import os

def read_and_display_file(filename):
  try:
    with open(filename, 'r') as f:
      content = f.read()
      print(f"Content of '{filename}': ")
      print(content)

      words = content.split()
      word_count = len(words)
      print(f"Number of words in the file: {word_count}")

  except FileNotFoundError:
    print(f"Error: File '{filename}' not found.")
  except Exception as e:
    print(f"An error occurred: {e}")

filename = input("Enter the name of the file to read: ")

read_and_display_file(filename)
```

c.	Python code for image processing

```python
from google.colab import files
uploaded = files.upload()
filename = list(uploaded.keys())[0]
print(f'Uploaded file: {filename}')

import cv2
import matplotlib.pyplot as plt
img = cv2.imread(filename)

# OpenCV reads images in BGR format, convert it to RGB for displaying correctly in matplotlib
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
flipped_img = cv2.flip(img_rgb, 1)  # Flip the image horizontally (left-right)
gray_img = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2GRAY) # Convert image to grayscale
plt.figure(figsize=(15, 5)) # Display the images

plt.subplot(1, 3, 1)
plt.imshow(img_rgb)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(flipped_img)
plt.title('Flipped Image')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(gray_img, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')

plt.show()
```

d. python program to upload a video and break it into frames

```python
import cv2
from google.colab import files
import os

uploaded = files.upload()

for filename in uploaded.keys():
  # Create a directory to save the frames
  output_dir = f"{filename}_frames"
  os.makedirs(output_dir, exist_ok=True)

  # Open the video file
  cap = cv2.VideoCapture(filename)

  if not cap.isOpened():
    print(f"Error: Could not open video file {filename}")
    continue

  frame_count = 0
  while True:
    ret, frame = cap.read()
    if not ret:
      break

    # Save the frame as an image file
    frame_filename = os.path.join(output_dir, f"frame_{frame_count:04d}.png")
    cv2.imwrite(frame_filename, frame)

    frame_count += 1

  # Release the video capture object
  cap.release()

  print(f"Successfully extracted {frame_count} frames from '{filename}' and saved to '{output_dir}'.")
```
